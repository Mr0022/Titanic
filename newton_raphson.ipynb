```r
# Load necessary library
library(readr)
library(dplyr)
# Read the CSV file
data = read_csv("https://raw.githubusercontent.com/Mr0022/Titanic/refs/heads/main/train.csv")
data
```

```r
data= subset(data, select = -c(PassengerId , Name ,Parch ,Ticket ,Cabin , Embarked,Fare, SibSp ))
data= data[, c(2:ncol(data), 1)]
# Assume you have a dataframe called df
# Remove rows with NA values
df_clean = na.omit(data)
df_clean
```

```r
# Load necessary libraries
library(car)

# Perform Levene's Test on Age by Sex and Pclass
levene_test <- leveneTest(Age ~ as.factor(Sex) * as.factor(Pclass), data = df_clean)

# Display the result
print(levene_test)
```

```r
# Load necessary libraries
library(ggplot2)

# Plot the data
ggplot(df_clean, aes(x = interaction(Sex, Pclass), y = Age)) +
  geom_boxplot() +
  labs(title = "Levene's Test on Age by Sex and Pclass",
       x = "Sex and Pclass",
       y = "Age") +
  theme_minimal()
```

```r
grouped_data <- df_clean %>%
  group_by(Sex, Pclass) %>%
  summarise(
    AvgAge = mean(Age), # Calculate the average age for each group
    n = n(), # Total number of passengers (n)
    y = sum(Survived) # Total number of survivors (y)
  )
grouped_data
```

```r
# Load the dataset

df <- df_clean
df$Sex <- ifelse(df$Sex == "male", 1, 0)


# Equation parameters
intercept <- 2.1998
coef_gender <- -3.6862
coef_class <- -0.6346
coef_avg_age <- 0.0314

# Define the logistic function
logistic <- function(log_odds) {
  return(1 / (1 + exp(-log_odds)))
}
# Convert 'Sex' column to numeric (male = 1, female = 0)
# Calculate log odds
df <- df %>%
  mutate(
    log_odds = intercept + coef_gender * Sex + coef_class * Pclass + coef_avg_age * Age,
    predicted_probability = logistic(log_odds)
  )

# Plot
ggplot(df, aes(x = log_odds, y = predicted_probability)) +
  geom_point(color = 'blue') +
  labs(
    title = "Logistic Regression: Predicted Probability vs. Log Odds",
    x = "Log Odds",
    y = "Predicted Probability"
  ) +
  theme_minimal()
```
