```r
# Load necessary library
library(readr)
library(dplyr)
# Read the CSV file
data = read_csv("https://raw.githubusercontent.com/Mr0022/Titanic/refs/heads/main/train.csv")
data
```

```r
data= subset(data, select = -c(PassengerId , Name ,Parch ,Ticket ,Cabin , Embarked,Fare, SibSp ))
data= data[, c(2:ncol(data), 1)]
# Assume you have a dataframe called df
# Remove rows with NA values
df_clean = na.omit(data)
df_clean
```

```r
grouped_data <- df_clean %>%
  group_by(Sex, Pclass) %>%
  summarise(
    AvgAge = mean(Age), # Calculate the average age for each group
    n = n(), # Total number of passengers (n)
    y = sum(Survived) # Total number of survivors (y)
  )
grouped_data
```

```r
library(ggplot2)
library(dplyr)

# Define logistic function to compute probability from log-odds
logistic <- function(log_odds) {
  return(1 / (1 + exp(-log_odds)))
}

# Equation parameters
intercept = -0.5164
coef_gender = -2.9786
coef_class = -0.349
coef_avg_age = 0.0902

# Create a dataframe for plotting, varying gender, class, and age
plot_data <- expand.grid(
  Gender = c(0, 1),  # 0 = Female, 1 = Male
  Class = c(1, 2, 3),  # Passenger classes
  AvgAge = seq(0, 80, by = 5)  # Age range from 0 to 80
)

# Calculate the log-odds and probability for each combination
plot_data <- plot_data %>%
  mutate(
    log_odds = intercept + coef_gender * Gender + coef_class * Class + coef_avg_age * AvgAge,
    Probability = logistic(log_odds),
    GenderLabel = ifelse(Gender == 1, "Male", "Female")  # Add a label for gender
  )

# Plot the probability of survival against age using points, colored by class and shaped by gender
ggplot(plot_data, aes(x = AvgAge, y = Probability, color = as.factor(Class), shape = as.factor(Gender))) +
  geom_point(size = 4) +
  labs(
    title = "Probability of Survival by Age, Gender, and Class",
    x = "Average Age",
    y = "Probability of Survival",
    color = "Class",
    shape = "Gender"
  ) +
  scale_shape_manual(values = c(16, 17), labels = c("Female", "Male")) +  # Different shapes for genders
  theme_minimal()

```
